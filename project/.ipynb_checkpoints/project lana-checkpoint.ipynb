```python
import csv

f = open('environmental.csv', 'rb')
reader = csv.reader(f)
rownum = 0
temperature = []
ozone = []

for row in reader:
    if rownum != 0:
        a=row[0]
        b=float(a)
        c=row[2]
        d=int(c)
        temperature.append(b)
        ozone.append(d)
    rownum += 1


print temperature
print ozone

f.close()





```

```python


from pylab import *
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('Temperature AND Ozone')
ax.set_xlabel('Temperature(c)')
ax.set_ylabel('Ozone')
ax.scatter(temperature, ozone)
fig.savefig("temp_with_titles.png")
```

When the temperature goes up ; the Ozone layer gets thicker but if it gets too much hot or too much cold the Ozone layer goes thinner and we dont want that to happen !
